**Dropping stuff**
# opposed to the concept of selection, we have "dropping". instead of pointing out which values you'd like to slect you could point which ones you'd like to drop

```python
df.drop('Canada')
```

```python
df.drop(['Canada','Japan'])
```

```python
df.drop(columns=['Population','HDI'])
```

```python
df.drop(['Italy','Canada'], axis=0)
```

```python
df.drop(['Population','HDI'], axis=1)
```

```python
df.drop(['Population','HDI'], axis='columns')
```

```python
df.drop(['Canada', 'Germany'], axis='rows')
```

all these drop methods return a new DataFrame. if you'd like to modify it "in place", you can use the inplace attribute (there's an example below).

**operations**

```python
df[['Population', 'GDP']]
```

operations with series work at a column level, broadcasting down the rows (which can be counter intutive).

```python
crisis = pd.Series([-1_000_000, -0.3], index=['GDP', 'HDI'])
crisis
```

```python
df[['GDP', 'HDI']]
```

```python
df[['GDP', 'HDI']] + crisis
```

# modifying dataframe

you can addcolumn, or replace values for columns without issues

**adding a new column**

```python
s = pd.Series(
    ['French', 'German', 'Italian'],
    index=['France', 'Germany', 'Italy'],
    name='Language'
)
```

```python
s
```

```python
df['language'] = s
```

```python
df
```

# Replacing values per column

```python
df['language'] = 'English'
```

```python
df
```

# Renaming column

```python
df.rename(
    columns={
        'HDI': 'Human Development Index',
        'Anual Popcorn Consumption': 'APC'
    }, index={
        'United States': 'USA',
        'United Kingdom': 'UK',
        'Argentina': 'AR'
    })
```

```python
df.rename(index=str.upper)
```

```python
df.rename(index=lambda x: x.lower())
```
